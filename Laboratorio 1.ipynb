LIBRERÍA **scikit-image**


**skimage** con el módulo **io**

Esta librería nos ayuda a leer y mostrar imágenes desde nuestro archivo de datos. En este ejemplo vamos a mostrar una imagen que se encuentra en el archivo **animal_data_img.csv** (tomamos la primera imagen), debemos tomar en cuenta que los nombres deben ser los mismos al momento de codificar, ya que de no ser así, nos reflejará un error.

```python
from skimage import io
```

**Ejemplo 1** (librería skimage)

Asignamos una variable **image** a la imagen e ingresamos el nombre como tenemos en nuestro archivo de datos.

```python
image=io.imread("1633802635025_Lion Shepherd Puppies for sale.jpg")

plt.imshow(image)
```

**Ejemplo 2** (librería skimage)

Con esta librería es posible modificar el color de la imagen, añadiendo el módulo **color** a la librería. 

En este ejemplo vamos a tomar la misma imagen y la convertiremos a escala de grises.

```python
from skimage import io,color

img = io.imread("1633802635025_Lion Shepherd Puppies for sale.jpg")
img_gris = color.rgb2gray(img)
io.imshow(img_gris)
io.show()
```

**Ejemplo 3** (librería skimage)

Con esta librería podemos unir dos imágenes en un mismo panel utilizando la función **subplot**.

Para este ejemplo uniremos las dos imágenes anteriores.

```python
import matplotlib.pyplot as plt
from skimage import io,color

img = io.imread("1633802635025_Lion Shepherd Puppies for sale.jpg")
img_gris = color.rgb2gray(img)
plt.subplot(211)
io.imshow(img)
plt.subplot(212)
io.imshow(img_gris)
io.show()
```
